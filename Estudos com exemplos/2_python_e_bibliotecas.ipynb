## Exemplo 2 - Atribuição de Variáveis

```python
cars = 100
spaceInACar = 4.0
drivers = 30
passengers = 90
carsNotDriven = cars - drivers
carsDriven = drivers
carpoolCapacity = carsDriven * spaceInACar
averagePassengersPerCar = passengers / carsDriven

print("There are", cars, "cars available.")
print("There are only", drivers, "drivers available.")
print("There will be", carsNotDriven, "empty cars today.")
print("We can transport", carpoolCapacity, "people today.")
print("We have", passengers, "to carpool today.")
print("We need to put about", averagePassengersPerCar, "in each car.")
print(10/5)
```

## Exemplo 5 - Estruturas Condicionais

```python
people = 30
cars = 40
trucks = 15

if cars > people:
    print("We should take the cars.")
elif cars < people:
    print("We should not take the cars.")
else:
    print("We can't decide.")

if trucks > cars:
    print("That's too many trucks.")
elif trucks < cars:
    print("Maybe we could take the trucks.")
else:
    print("We still can't decide.")

if people > trucks:
    print("Alright, let's just take the trucks.")
else:
    print("Fine, let's stay home then.")
```

## Exemplo 6 - Laços de Repetição

```python
i = 0
numbers = []

while i < 6:
    print(f"At the top i is {i}")
    numbers.append(i)

    i = i + 1
    print("Numbers now: ", numbers)
    print(f"At the bottom i is {i}")


print("The numbers: ")

for num in numbers:
  print(num)
```

## Exemplo 7 - Listas e Strings

```python
the_count = [1,2,3,4,5]
fruits = ['apples', 'oranges', 'pears', 'apricots']
change = [1, 'pennies', 2, 'dimes', 3, 'quarters']
lista = []

# esse primeiro tipo de loop for percorre uma lista
for number in the_count:
    print(f"This is count {number}")

# mesma coisa que o código acima
for fruit in fruits:
    print(f"A fruit of type: {fruit}")

# também podemos percorer listas mistas
# perceba que temos que usar um {} uma vez que não sabemos o que há nela
for i in change
  print(f"I got {i}")

# também podemos construir listas, primeiro comece com uma vazia
elements = []

# então use a função range para fazer a contage de 0 a 5
for i in range(0, 6):
  print(f"Adding {i} to the list.")
  # append é uma função que lista faz
  elements.append(i)

# agora podemos imprimir a lista
for i in elements:
  print(f"Element was: {i}")
```

## Exercício 1: Elaborar um programa que calcule e apresente o valor do voulme de uma caixa retangular, utilizando a fórmula VOLUME = COMPRIMENTO * LARGURA * ALTURA

```python
COMPRIMENTO = float(input("Digite o comprimento da caixa: "))
LARGURA = float(input("Digite a largura da caixa: "))
ALTURA = float(input("Digite a altura da caixa: "))

VOLUME = COMPRIMENTO * LARGURA * ALTURA

print(f"Uma caixa com comprimento {COMPRIMENTO}, largura {LARGURA} e altura {ALTURA} tem um volume de {VOLUME} cm^3")
```

## Exercício 2: Contar o número de palavras em um texto

```python
texto = input("Digite um texto: ")
palavras = texto.split()
numero_palavras = len(palavras)
print(f"O texto tem {numero_palavras} palavras.")
```

## Exercício 3: Buscar e substituir palavras em um texto

```python
texto = input("Digite um texto: ")
palavra_antiga = input("Digite a palavra antiga: ")
nova_palavra = input("Digite a nova palavra: ")
novo_texto = texto.replace(palavra_antiga, nova_palavra)
print(novo_texto)
```

## Exercício 4: Separar frases e iterar sobre elas

```python
paragrafo = input("Digite um paragrafo: ")
frases = paragrafo.split(".")
for i, frase in enumerate(frases):
  if frase.strip() != "":
    print(f"Frase {i+1}: {frase.strip()}")
```

## Exercício 5: Ler três valores inteiros representados pelas variáveis A,B e C e apresentar os valores lidos dispostos em ordem crescente.

```python
A = int(input("Digite o valor de A: "))
B = int(input("Digite o valor de B: "))
C = int(input("Digite o valor de C: "))

if A < B and A < C:
  if B < C:
    print(f"A ordem crescente é {A}, {B} e {C}")
  else:
    print(f"A ordem crescente é {A}, {C} e {B}")
```

## Exercício 6: Fazer a leitura de quatro valores numéricos inteiros representados pelas variáveis A,B,C e D. Apresentar os valores que sejam divisíveis por 2 e 3

```python
A = int(input("Digite o valor de A: "))
B = int(input("Digite o valor de B: "))
C = int(input("Digite o valor de C: "))
D = int(input("Digite o valor de D: "))

lista = [A, B, C, D]
for num in lista:
    if num % 2 == 0 and num % 3 == 0:
        print(num)
```
